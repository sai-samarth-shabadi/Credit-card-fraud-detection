#importing the required dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
#load the dataset
df = pd.read_csv('creditcard.csv')
```

```python
#first 5 rows of data
df.head()
```

```python
#Gathring information about the data
df.info()
```

```python
#checking missing values
df.isnull().sum()
```

```python
# distribution of legitimate and fraudlent transactions
df['Class'].value_counts()
```

#This dataset is highly imbalanced

#0--> legitimate transactions
#1--> Fraudlent transactions

```python
#separating the data for analysis
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
#satistical measures of the data
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

```python
# Compare the values for both transactionns
df.groupby('Class').mean()
```

#Under-Sampling

#Build a sample dataset containing similar distribution of normal and fraudlent transactions

#Number of fraudlent transactions --> 492

```python
legit_sample = legit.sample(n= 492)
```

```python
new_df = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

#Splitting the Features and Target Columns

```python
x = new_df.drop(columns = 'Class', axis = 1)
y = new_df['Class']
```

```python
#splitting train and test data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, stratify = y, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

#Creating a Pipeline

#Pipeline consist of Standard scaler and logistic Regression

```python
model = make_pipeline(StandardScaler(), LogisticRegression())
```

```python
#Trainiing the model
model.fit(x_train, y_train)
```

#Model Evaluation

#Accuracy Score

```python
#Accuracy on training data
y_train_pred = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train_pred, y_train)
```

```python
print("Accuracy Score on Training Data : ", training_data_accuracy)
```

```python
#Accuracy on testing data
y_test_pred = model.predict(x_test)
testing_data_accuracy = accuracy_score(y_test_pred, y_test)
```

```python
print("Accuracy Score on Testing Data : ", testing_data_accuracy)
```
